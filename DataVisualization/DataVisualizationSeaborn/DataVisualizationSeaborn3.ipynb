Data Visualization with Seaborn part 3!

```python
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

Using hue and colormapping to show additional information.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

Changing shape of point and point size to better distinguish between male and female points.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',
           markers=['o','v'],scatter_kws={'s':100})
```

Using row and col attributes to produce multiple graphs that can be easily compared.

```python
sns.lmplot(x="total_bill", y="tip", row="sex", col="time",data=tips)
```

Adjusting the sizing of graphs using aspect and height attributes.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,height=8)
```

```python
import matplotlib.pyplot as plt
```

We can adjust the style and sizing in the following way as well.

```python
sns.set_style('white')
plt.figure(figsize=(12,6))
sns.countplot(x='sex',data=tips)
```
